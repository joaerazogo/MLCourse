```python
def getNeighbors(trainingSet, labels,
                 testInstance, k,
                 distance = distance):

    """
    la función getNeighbors calcula la listra de los k vecinos más cercanos
    de una instancia 'testInstance'
    la lista de los vecinos contiene 3 tuplas con (index, dist, label)
    donde:
    index: es el indice del conjunto de entrenamiento
    dist: es la distancia entre testInstance (instancia de prueba) y la instancia del conjunto de entrenamiento trainingSet[index]
    distance: es una referencia a una función usada para calcular las distancias
    """
    distances = []
    for index in range(len(trainingSet)):
        dist = distance(testInstance, trainingSet[index])
        distances.append((trainingSet[index], dist, labels[index]))
    
    distances.sort(key=lambda x: x[1])
    neighbors = distances[:k]
    return neighbors
```

```python
def vote_harmonic_weigths(neighbors, allResults=True):
    classCounter = Counter()
    numberOfNeighbors = len(neighbors)
    for index in range(numberOfNeighbors):
        classCounter[neighbors[index][2]] += 1/(index+1)
        pass
    labels, votes = zip(*classCounter.most_common())
    winner = classCounter.most_common(1)[0][0]
    votes4winner = classCounter.most_common(1)[0][1]
    if allResults:
        total = sum(classCounter.values(), 0.0)
        for key in classCounter:
            classCounter[key] /= total
        return winner, classCounter.most_common()
    else:
        return winner, votes4winner/sum(votes)
        pass
    pass
```

```python
def vote_distance_weights(neighbors, all_results=True):
    class_counter = Counter()
    number_of_neighbors = len(neighbors)
    for index in range(number_of_neighbors):
        dist = neighbors[index][1]
        label = neighbors[index][2]
        class_counter[label] += 1 / (dist**2 + 1)
    labels, votes = zip(*class_counter.most_common())
    print(class_counter)
    #print(labels, votes)
    winner = class_counter.most_common(1)[0][0]
    votes4winner = class_counter.most_common(1)[0][1]
    if all_results:
        total = sum(class_counter.values(), 0.0)
        for key in class_counter:
             class_counter[key] /= total
        return winner, class_counter.most_common()
    else:
        return winner, votes4winner / sum(votes)
```

```python
def vecinosCercanos(dataTest,
                    dataSwitch,
                    classifierSwitch,
                    k,
                    distance=distance):
    
    distanceInstances = []
    for dataIndex in range(len(dataSwitch)):
        dist = distance(dataTest, [dataSwitch['Lag1'][dataIndex],dataSwitch['Lag2'][dataIndex]])
        distanceInstances.append([dataTest, dist, [dataSwitch['Lag1'][dataIndex],dataSwitch['Lag2'][dataIndex]], classifierSwitch['Direction'][dataIndex]])
    
    distanceInstances.sort(key=lambda vector: vector[1])
    vecinos = distanceInstances[:k]
    return vecinos
    pass
```
