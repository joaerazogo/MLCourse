```python
import pandas as pd
import io
import requests
url="https://raw.githubusercontent.com/madmashup/targeted-marketing-predictive-engine/master/banking.csv"
request = requests.get(url).content
dataset_marketing = pd.read_csv(io.StringIO(request.decode('utf-8')))
```

```python
dataset_marketing
```

```python
dataset_marketing['education'].unique()
```

```python
# Se realiza la consulta por basic.9y, basic.6y, basic.4y para reemplazar estos valores por Basic
dataset_marketing['education'] = np.where(dataset_marketing['education'] =='basic.9y', 'Basic', dataset_marketing['education'])
dataset_marketing['education'] = np.where(dataset_marketing['education'] =='basic.6y', 'Basic', dataset_marketing['education'])
dataset_marketing['education'] = np.where(dataset_marketing['education'] =='basic.4y', 'Basic', dataset_marketing['education'])
```

```python
dataset_marketing['education'].unique()
```

```python
dataset_marketing['y'].value_counts()
```

```python
sns.countplot(x='y', data=dataset_marketing, palette='hls')
plt.show()
_plt.savefig('count_plot')
```

```python
count_no_sub = len(dataset_marketing[dataset_marketing['y'] == 0])
count_sub = len(dataset_marketing[dataset_marketing['y'] == 1])
pct_of_no_sub = count_no_sub/(count_no_sub + count_sub)
print("percentage of no subcription is", pct_of_no_sub*100)
pct_of_sub = count_sub/(count_no_sub + count_sub)
print("percentage of subcription is", pct_of_sub*100)
```

```python
#%matplotlib.inline
pd.crosstab(dataset_marketing.job,dataset_marketing.y).plot(kind='bar')
plt.title('Purchase Frequency for Job Title')
plt.xlabel('Job')
plt.ylabel('Frequency of Purchase')
plt.savefig('purchase_fre_job')
```

```python
table = pd.crosstab(dataset_marketing.marital,dataset_marketing.y)
table.div(table.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True)
plt.title('Stacked bar chart of marital status vs purchase')
plt.xlabel('Marital status')
plt.ylabel('Portion of customers')
plt.savefig('marital vs pur stack')
```
