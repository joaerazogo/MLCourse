```python
mainPath = '/home/jose/Downloads/python-ml-course/datasets/'
filepath = 'customer-churn-model/Customer Churn Columns.csv'
data2_Customer = pd.read_csv(os.path.join(mainPath, filepath), header=None, names=data2_columns)
data2_Customer.columns.values
```

```python
data3 = open(mainPath + filepath_Customer, 'r')
```

```python
cols = data3.readline().strip().split(',')
cols
```

```python
ncols = len(cols)
ncols
```

```python
counter = 0
main_dict = {}

for col in cols:
    main_dict[col] = []
```

```python
for line in data3:
    values = line.strip().split(',')
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1
    
print('El dataset tiene %d filas y %d columnas '%(counter, ncols))
```

```python
dataSet3 = pd.DataFrame(main_dict)
dataSet3
```

```python
infile = mainPath + filepath_Customer
outfile = mainPath + 'customer-churn-model/Tab Customer Churn Model.txt'
```

```python
with open(infile, 'r') as infile1:
    with open(outfile, 'w') as outfile1:
        for line in infile1:
            fields = line.strip().split(',')
            outfile1.write('\t'.join(fields))
            outfile1.write('\n')
```

```python
dataFrame4 = pd.read_csv(outfile, sep='\t')
dataFrame4
```

# Leer los datos desde una url

```python
medals_url = 'http://winterolympicsmedals.com/medals.csv'
```

```python
medals_data = pd.read_csv(medals_url)
medals_data
```

```python
import csv
import urllib3

http = urllib3.PoolManager()
r = http.request('GET', medals_url)
r.status
df_urllib = r.data
df_urllib
```

```python
import csv

data_medals = csv.reader(df_urllib)
data_medals
```

```python
for row in data_medals:
    print(row)
```
