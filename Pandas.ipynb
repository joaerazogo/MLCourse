```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

periods = 7 * 24 * 60
tidx = pd.date_range('2016-07-01', periods=periods, freq='T')
np.random.seed([3,1415]) # This will pick a number of normally distributed random numbers where the number is specified by periods
data = np.random.randn(periods)

ts = pd.Series(data=data, index=tidx, name='HelloTimeSeries')
ts.resample('15T').agg(['min', 'mean', 'max', 'std']).plot()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 4))
for i, freq in enumerate(['15T', '30T', '1H']):
    ts.resample(freq).agg(['min', 'mean', 'max', 'std']).plot(ax=axes[i], title=freq)
```
